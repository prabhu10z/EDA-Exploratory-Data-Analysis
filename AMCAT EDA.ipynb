## Univariate Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,6))
plt.boxplot(data['Salary'])
plt.title('Boxplot for Column Salary')
plt.xlabel('Column')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

#### Observation : The boxplot visualizes the distribution of salary values, indicating a wide range of salaries with several outliers towards the higher end. The median salary lies within the lower quartile, suggesting a potential skewness towards lower incomes, while the upper whisker denotes considerable variability in higher salary ranges.

```python
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

sns.histplot(data['English'], kde=True, ax=axs[0], color='blue')
axs[0].set_title('Distribution of English Scores')
axs[0].set_xlabel('English Scores')
axs[0].set_ylabel('Frequency')


sns.histplot(data['ComputerScience'], kde=True, ax=axs[1], color='pink')
axs[1].set_title('Distribution of ComputerScience Scores')
axs[1].set_xlabel('ComputerScience Scores')
axs[1].set_ylabel('Frequency')

sns.histplot(data['MechanicalEngg'], kde=True, ax=axs[2], color='orange')
axs[2].set_title('Distribution of MechanicalEngg  Scores')
axs[2].set_xlabel('MechanicalEngg   Scores')
axs[2].set_ylabel('Frequency')


plt.tight_layout()
plt.show()
```

#### Observation : The histograms represents the 'English', 'ComputerScience', and 'MechanicalEngg  ' columns, showcasing the distribution of scores for each subject. Kernel density estimates (KDE) are included to provide smoother representations of the distributions. These visualizations offer insights into the spread and central tendencies of scores in each subject.

```python
plt.figure(figsize=(8, 6))
plt.boxplot(data['collegeGPA'])
plt.title('Boxplot for collegeGPA')
plt.xlabel('collegeGPA')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

#### Observation: The boxplot illustrates the distribution of college GPAs, with a median value close to the center of the interquartile range. There are a few outliers on the lower end and some are in upper range, suggesting a relatively normal distribution with some students performing below the median GPA.

```python
import matplotlib.pyplot as plt
import seaborn as sns
numerical_cols = ['Logical','ComputerProgramming','ElectronicsAndSemicon']
plt.figure(figsize=(10, 6))
sns.boxplot(data=data[numerical_cols], palette='Set3')
plt.title('Boxplot for Multiple Numerical Columns')
plt.xlabel('Columns')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.show()
```

#### Observation:This combined boxplot visualizes the distribution of values across multiple numerical columns, including 'Logical', 'MechanicalEngg', and 'ComputerProgramming'. Each box represents the spread of values within its respective column, with whiskers extending to the minimum and maximum values and a line indicating the median. 'Logical' and 'ComputerProgramming' columns exhibit notable variability, while 'ElectronicsAndSemicon' displays a relatively tighter distribution. The plot provides a comparative overview of these numerical variables' distributions, aiding in identifying potential patterns or outliers across the dataset.

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 6))
sns.histplot(data['MechanicalEngg'], kde=True, ax=axs[0], color='maroon')
axs[0].set_title('PDF of MechanicalEngg')
axs[0].set_xlabel('Value')
axs[0].set_ylabel('Density')

sns.histplot(data['ElectricalEngg'], kde=True, ax=axs[1], color='yellow')
axs[1].set_title('PDF of ElectricalEngg')
axs[1].set_xlabel('Value')
axs[1].set_ylabel('Density')

plt.tight_layout()
plt.show()
```

#### Observation:This visualization depicts the probability density function (PDF) of the MechanicalEngg column and ElectricalEngg, showcasing a distribution with a peak density around a certain value. The kernel density estimate (KDE) overlay provides a smoothed representation of the distribution's shape, aiding in identifying its central tendency and spread.

```python
top_JobCity= data['JobCity'].value_counts().nlargest(20)

plt.figure(figsize=(20, 10))
top_JobCity.plot(kind='bar', color='green')
plt.title('Frequency Distribution of Categories')
plt.xlabel('JobCity')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

The 'JobCity' category 'Frequency' demonstrates a varied distribution, with certain designations appearing more frequently than others, while the horizontal orientation of the x-axis labels aids readability for a larger number of categories. from above bar we can say that the software engineer has the highest frequency followed by software developer.

```python
sns.kdeplot(data,fill=True)
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Kernel Density Estimation (KDE) Plot')
plt.show()
```

#### Observation:The Kernel Density Estimation (KDE) plot visually represents the distribution of multiple variables in the DataFrame df, highlighting the density of data points across their respective ranges. The shaded areas emphasize regions with higher data density, providing insights into the overall distribution pattern and concentration of values within the dataset.

## bivariate analysis

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['10percentage'], data['12percentage'],color="violet")
plt.title('Scatter Plot')
plt.xlabel('10th percentage')
plt.ylabel('12th percentage')
plt.grid(True)
plt.show()
```

#### Observation:This scatter plot visualizes the relationship between the '10th percentage' and '12th percentage' columns. The points are scattered across the plot, indicating the distribution and correlation between the two variables as we can say that 10th percentages between 60 to 95 and 12t percentages of 60 to 99 has the highest density.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='Salary', data=data)
plt.title('Salary Distribution by Gender')
plt.show()
```

#### Observation:Look for significant differences in the median salary between genders. A higher or lower median might indicate gender disparities. as we can observes that the gender male as the highest salary outliers as compare to the female

```python
plt.figure(figsize=(12, 8))
top_designations = data['Designation'].value_counts().index[:10]
sns.boxplot(x='Salary', y='Designation', data=data[data['Designation'].isin(top_designations)], orient='h')
plt.title('Salary by Designation for Top 10 Designations')
plt.show()
```

#### Observation:This will help identify if certain designations tend to have higher salaries.This horizontal boxplot compares the distribution of salaries across the top 10 most frequent designations. Each box represents the salary range for a specific designation, providing insights into the salary distribution within different job roles. Higher variability in salary is observed among top designations, indicating potential differences in compensation based on job titles.

```python
numerical_cols = ['Salary', '10percentage', '12percentage', 'collegeGPA', 'English', 'Logical', 'Quant', 'conscientiousness', 'agreeableness', 'extraversion', 'nueroticism', 'openess_to_experience']

correlation_matrix = data[numerical_cols].corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```

#### Observation:The correlation heatmap reveals the pairwise correlations between numerical variables in the dataset. Strong positive correlations are observed, notably between '10percentage' and '12percentage', 'Salary' and 'collegeGPA', suggesting a relationship between academic performance metrics and salary. Conversely, 'Salary' exhibits weak correlations with personality traits like 'conscientiousness' and 'nueroticism', indicating that job salary may not strongly correlate with these personality attributes

```python
cross_tab = pd.crosstab(data['Gender'], data['Degree'])
cross_tab.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Stacked Bar Plot between Gender and Degree')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

#### Observation:The stacked bar plot illustrates the distribution of degree types across genders. It highlights the gender-wise breakdown of degree categories, showcasing the relative frequencies of each degree type among males and females. The plot reveals the gender disparities in educational backgrounds, with certain degrees being more prevalent among one gender compared to the other. here we can observe that number of males has the has degree in B.Tech/BE and the male who are completed MCA has second highest number, in female also most of the female completed degree in B.Tech/B.E and few has done MCA andd very few has completed M.SC.(Tech.)

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.stripplot(x='Gender', y='Specialization', data=data)
plt.title('Strip Plot between Gender and Specialization')
plt.xlabel('Gender')
plt.ylabel('Specialization')
plt.show()

# Bar Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Degree', y='Salary', data=data, estimator=sum)
plt.title('Bar Plot between Degree and Total Salary')
plt.xlabel('Degree')
plt.ylabel('Total Salary')
plt.show()
```

#### Observation:The strip plot between gender and specialization displays the distribution of different specializations across genders. It reveals the clustering of various specializations based on gender, highlighting potential gender-based preferences or trends in educational pursuits.

#### The bar plot illustrates the total salary aggregated by degree types, providing insights into the overall earning potential associated with different degrees. It enables comparison of the total salary earned across various degree categories, aiding in identifying degrees that contribute more significantly to overall earnings.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Salary', y='collegeGPA', data=data)
plt.title('Scatter Plot between Salary and collegeGPA')
plt.xlabel('Salary')
plt.ylabel('collegeGPA')
plt.show()

numerical_columns = ['Salary', 'collegeGPA', 'English', 'Logical', 'Quant']
sns.pairplot(data[numerical_columns])
plt.show()
```

#### Observation:The scatter plot between salary and college GPA reveals a positive correlation between these two variables, suggesting that individuals with higher GPAs tend to earn higher salaries.

- The pair plot provides a comprehensive overview of pairwise relationships between numerical variables. Strong positive correlations are observed between 'Salary' and 'collegeGPA', 'English' and 'Logical', indicating potential relationships between academic performance metrics. Additionally, the plot highlights distributions and correlations among other numerical variables, aiding in identifying patterns and potential insights within the dataset.
